```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: iPhone sales (in millions) over the years
years = np.array([2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])
sales = np.array([231, 211, 217, 218, 196, 200, 240, 225, 235])  # Hypothetical data

plt.plot(years, sales, marker='o', linestyle='-', color='b', label="iPhone Sales")

# Labels and title
plt.xlabel("Year")
plt.ylabel("Sales (millions)")
plt.title("iPhone Sales Over the Years")
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: iPhone models and their launch prices in USD
models = ["iPhone 6", "iPhone 7", "iPhone 8", "iPhone X", "iPhone 11", "iPhone 12"]
prices = np.array([649, 699, 699, 999, 699, 799])  # Hypothetical prices

plt.plot(models, prices, marker='o', linestyle='-', color='g', label="iPhone Prices")

# Labels and title
plt.xlabel("iPhone Models")
plt.ylabel("Launch Price (USD)")
plt.title("iPhone Launch Prices Over Different Models")
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: iPhone models and their battery capacities in mAh
models = np.array(["iPhone X", "iPhone 11", "iPhone 12", "iPhone 13"])
battery_capacity = np.array([2716, 3110, 2815, 3240])  # Hypothetical values in mAh

plt.barh(models, battery_capacity, color='skyblue')

# Labels and title
plt.xlabel("Battery Capacity (mAh)")
plt.ylabel("iPhone Models")
plt.title("iPhone Battery Capacity Comparison")

# Show the plot
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating sample resale price data (normal distribution)
resale_prices = np.random.normal(500, 100, 250)  # Mean price = $500, Std dev = $100

plt.hist(resale_prices, bins=15, color='skyblue', edgecolor='black')

# Labels and title
plt.xlabel("Resale Price (USD)")
plt.ylabel("Frequency")
plt.title("Distribution of iPhone Resale Prices")

# Show the plot
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: Popularity of different iPhone storage variants
labels = ["128GB", "256GB", "512GB", "1TB"]
storage_distribution = np.array([40, 35, 20, 5])  # Hypothetical percentages

plt.pie(storage_distribution, labels=labels, autopct='%1.1f%%', colors=["blue", "orange", "green", "red"])

# Title
plt.title("iPhone Storage Variant Distribution")

# Show the plot
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white")

# Generating sample sales data (number of units sold in millions)
rs = np.random.RandomState(25)
sales_data = rs.normal(loc=50, scale=15, size=200)  # Mean = 50 million, Std dev = 15 million

# Plot histogram with KDE
sns.histplot(sales_data, kde=True, color="g", bins=15)

# Labels and title
plt.xlabel("Units Sold (Millions)")
plt.ylabel("Frequency")
plt.title("Distribution of iPhone Sales (Units Sold)")

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data: iPhone benchmark scores across different models (hypothetical data)
data = {
    'iPhone Model': ['iPhone 12', 'iPhone 12', 'iPhone 13', 'iPhone 13', 'iPhone 14', 'iPhone 14', 'iPhone 15', 'iPhone 15'],
    'Benchmark Score': [1500, 1550, 1600, 1620, 1700, 1750, 1800, 1850],  # Hypothetical scores
}

# Create a DataFrame
df = sns.load_dataset('tips')  # Using seaborn's tips dataset for structure
df = pd.DataFrame(data)

# Create a boxplot to compare benchmark scores across different iPhone models
sns.boxplot(x='iPhone Model', y='Benchmark Score', data=df)

# Labels and title
plt.xlabel("iPhone Models")
plt.ylabel("Benchmark Scores")
plt.title("iPhone Performance Benchmark Scores Comparison")

# Show the plot
plt.show()
```
